```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt


# Define the symbolic variable
t = sp.Symbol('t')

alpha = 0.025*sp.sin(1.5*t) + 0.0481  # Define alpha (intrest rate)

# Integrate symbolically
alpha_int = sp.integrate(alpha, t) 

# Convert to a callable function
alpha_int_callable = sp.lambdify(t, alpha_int, 'numpy')

# define constants

T = 30
B0 = 200000

def rCalc(t):
    return (B0/T) * np.exp(alpha_int_callable(t))



T_array = np.linspace(0, 30, 1000)

r_Vals = rCalc(T_array)

plt.plot(T_array, r_Vals, color='g', label="Repayment rate")

# Adding labels and title
plt.xlabel('Time (Years)')
plt.ylabel('Repayment rate (Â£/Year)')
plt.title('Repayment rate as a function of time')
plt.legend()

# Show the plot
plt.tight_layout()
plt.grid(True)
plt.savefig('Repayment rate as a function of time (200k mortage over 30 years oscilating intrest.jpg')
plt.show()



    
```
